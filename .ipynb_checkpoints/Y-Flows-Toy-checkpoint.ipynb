```python
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import LogNorm
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import seaborn as sns
from geomloss import SamplesLoss


def sample_data(batch, dim, device):
    centers = np.array([[-3, 1.0],
                        [ 3, 1.0],
                        [ -1, -0.5],
                        [ 1, -0.5]
                       ])
    k = centers.shape[0]
    picks = np.random.randint(0, k, size=batch)
    pts = centers[picks] + 0.2 * np.random.randn(batch, 2)
    return torch.tensor(pts, dtype=torch.float32, device=device)

    
def sample_prior(batch, dim, device):
    return torch.randn(batch, dim, device=device) * 0.3 + torch.tensor([0.0, -7.0], device=device)


def visualize_source_to_target(model, step, dim, num_traj=100, steps=10, num_points=500, device = 'cuda:0'):
    x0_all = sample_prior(num_points, dim, device)   # sources
    x1_all = sample_data(num_points, dim, device)    # targets
    idx = np.random.choice(num_points, num_traj, replace=False)
    x0 = x0_all[idx]
    x1 = x1_all[idx]
    traj = [x0.detach().cpu().numpy()]
    x = x0

    # 3. Integrate ODE
    for k in range(steps):
        t = torch.full((num_traj,1), (k / steps), device=x.device)
        v = model(x, t)
        x = x + DT * v
        traj.append(x.detach().cpu().numpy())

    traj = np.stack(traj, axis=0)  # (steps+1, num_traj, 2)

    plt.figure(figsize=(8,8), dpi=150)
    ax = plt.gca()
    ax.set_facecolor("white")
    ax.grid(True, linestyle="--", alpha=0.5)

    cmap = plt.cm.inferno

    # Sources
    ax.scatter(x0_all[:,0].cpu(), x0_all[:,1].cpu(),
               c=[cmap(0.05)], s=20, edgecolor='white', linewidth=0.3,
               alpha=0.7, label='Source')

    # Targets
    ax.scatter(x1_all[:,0].cpu(), x1_all[:,1].cpu(),
               c=[cmap(0.80)], s=40, edgecolor='black', linewidth=0.8,
               alpha=0.9, label='Target')

    # Trajectories with continuous Inferno gradient
    colors = cmap(np.linspace(0.1, 0.95, traj.shape[0]))
    for i in range(num_traj):
        for k in range(traj.shape[0]-1):
            ax.plot(traj[k:k+2,i,0], traj[k:k+2,i,1],
                    color=colors[k], lw=1.5, alpha=0.9)

    ax.legend(frameon=True)
    ax.set_title(f"Source → Target Transport (iter {step})", fontsize=14, weight='bold')
    ax.set_xlim(-4, 4)
    ax.set_ylim(-8, 3)
    ax.set_aspect('equal', 'box')

    plt.tight_layout()
    plt.show()

    
def save_model(model, iteration, alpha, task='W'):
    """
    Saves the model state dictionary to a file.

    Args:
        model (nn.Module): The PyTorch model to save.
        iteration (int): The current training iteration number.
        alpha (float): The alpha hyperparameter, used in the folder name.
    """
    folder_name = f"./saved_models"
    os.makedirs(folder_name, exist_ok=True)
    file_path = os.path.join(folder_name, f"model_iter_{task}_{iteration}_{alpha}.pth")
    torch.save(model.state_dict(), file_path)
    print(f"✅ Model saved to {file_path}")


class TimeEmbedding(nn.Module):
    def __init__(self, dim_in, dim_out):
        super().__init__()
        self.lin = nn.Sequential(
            nn.Linear(dim_in, dim_out),
            nn.SiLU(),
            nn.Linear(dim_out, dim_out)
        )
    def forward(self, t):
        # t: (B,1)
        return self.lin(t)


class VelocityNet(nn.Module):
    def __init__(self, x_dim=2, time_emb=64, hidden=256):
        super().__init__()
        self.time_emb = TimeEmbedding(1, time_emb)
        self.net = nn.Sequential(
            nn.Linear(x_dim + time_emb, hidden),
            nn.SiLU(),
            nn.Linear(hidden, hidden),
            nn.SiLU(),
            nn.Linear(hidden, hidden),
            nn.SiLU(),
            nn.Linear(hidden, x_dim)
        )
    def forward(self, x, t):
        te = self.time_emb(t)  # (B, time_emb)
        inp = torch.cat([x, te], dim=-1)
        return self.net(inp)


def jacobian_frobenius_squared(v_fn, x, t):
    x = x.requires_grad_(True)
    t = t.requires_grad_(False)
    v = v_fn(x, t)  # (B, dim)
    B, D = v.shape
    norms = torch.zeros(B, device=x.device)

    # compute gradients of each output coordinate
    for i in range(D):
        grad_i = torch.autograd.grad(v[:, i].sum(), x, create_graph=True)[0]  # (B, D)
        norms += (grad_i ** 2).sum(dim=1)
    return norms  # (B,)


def integrate_ode(x0, model, alpha = 0.9, steps=10):
    x = x0
    batch = x.shape[0]
    total_potential = 0.0
    total_kinetic = 0.0
    for k in range(steps):
        t = torch.full((batch,1), (k / steps), device=x.device, dtype=x.dtype)  # t in [0,1)
        v = model(x, t)  # (B, dim)
        pot = (v.norm(dim=1) ** alpha)  # (B,)
        total_potential = total_potential + pot.mean()
        jac_sq = jacobian_frobenius_squared(model, x, t)  # (B,)
        total_kinetic = total_kinetic + jac_sq.mean()
        # Euler step
        x = x + DT * v
    return x, total_potential * DT, total_kinetic * DT
```

```python
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
DEVICE = 'cuda:0' if torch.cuda.is_available() else 'cpu'
DIM = 2            
ALPHA = 0.7
LR = 1e-3
STEPS = 10         
DT = 1.0 / STEPS
BATCH = 256
NUM_ITERS = 10000
LAMBDA_COHESION = 0.01
LAMBDA_SINKHORN = 5.0

model = VelocityNet(DIM, time_emb=32, hidden=128).to(DEVICE)
optim = torch.optim.Adam(model.parameters(), lr=LR)
sinkhorn_loss = SamplesLoss(loss="sinkhorn", p=2, blur=0.05, backend="tensorized")


print("Starting training...")
for it in range(1, NUM_ITERS + 1):
    optim.zero_grad()
    z = sample_prior(BATCH, DIM, DEVICE)
    xT, pot_term, kin_term = integrate_ode(z, model, alpha=ALPHA, steps=STEPS)
    x_data = sample_data(BATCH, DIM, DEVICE)
    
    loss_sink = sinkhorn_loss(xT, x_data)

    m_alpha_eps = pot_term +  LAMBDA_COHESION * kin_term
    loss = LAMBDA_SINKHORN * loss_sink + m_alpha_eps

    loss.backward()
    optim.step()

    if it % 500 == 0 or it == 1:
        print(f"Iter {it:6d}  loss={loss.item():.6f}  sinkhorn={loss_sink.item():.6f}  pot={pot_term.item():.6f} kin={kin_term.item():.6f}")
        visualize_source_to_target(model, it, dim=DIM, num_traj=50, steps=STEPS, device = DEVICE)
        save_model(model, it, ALPHA)
        print('model_saved')

print("Training finished.")
```
